**Data Pre-processing**

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    # Remove special characters and numbers
    text = re.sub(r'\W+', ' ', text)
    # Remove stopwords
    text = ' '.join([word for word in text.split() if word not in stop_words])
    # Lemmatization
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
    return text

movie_train_df['Description'] = movie_train_df['Description'].apply(preprocess_text)
movie_test_df['Description'] = movie_test_df['Description'].apply(preprocess_text)
```

```python
X = movie_train_df.drop(columns=['Genre'])
y = movie_train_df['Genre']
```

**Feature Extraction**

```python
# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))

# Fit and transform the train descriptions
X_train = tfidf.fit_transform(movie_train_df['Description'])

# Transform the test descriptions
X_test = tfidf.transform(movie_test_df['Description'])
```

**Handling Multilabel Classification**

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Split genres into lists
movie_train_df['Genre'] = movie_train_df['Genre'].apply(lambda x: x.split(','))

# Binarize the output labels
mlb = MultiLabelBinarizer()
y_train = mlb.fit_transform(movie_train_df['Genre'])

# Convert test solution genres into binary format
movie_test_solution_df['Genre'] = movie_test_solution_df['Genre'].apply(lambda x: x.split(','))
y_test = mlb.transform(movie_test_solution_df['Genre'])
```

**Model Training**

```python
# Initialize the model
model = OneVsRestClassifier(LogisticRegression(solver='liblinear', random_state=42))

# Train the model
model.fit(X_train, y_train)
```

**Model Evaluation**

```python
# Predict on the training data
y_train_pred = model.predict(X_train)

# Evaluate on training data
print("Training Set Evaluation:")
print(classification_report(y_train, y_train_pred, target_names=mlb.classes_))

# Predict on the test data
y_test_pred = model.predict(X_test)

# Evaluate on test data
print("Test Set Evaluation:")
print(classification_report(y_test, y_test_pred, target_names=mlb.classes_))
```

```python
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```
